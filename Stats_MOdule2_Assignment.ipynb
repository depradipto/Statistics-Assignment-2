 1. What is a random variable in probability theory?
 -  A random variable is a function that maps the outcomes of a random experiment to a set of real numbers.

 2. What are the types of random variables?
 - Discrete Random Variable: Can only take on a finite or countably infinite number of values (e.g., number of heads in a coin toss, number of cars passing a point in an hour).
 Continuous Random Variable: Can take on any value within a given range (e.g., height of a person, temperature).

 3. What is the difference between discrete and continuous distributions?
  - Discrete distributions represent data that can only take on specific, distinct values (like whole numbers), while continuous distributions represent data that can take any value within a range.

 4. What are probability distribution functions (PDF)?
  - A Probability Distribution Function (PDF) describes the likelihood of different outcomes in a random experiment. It provides a mathematical function that specifies the probability of a random variable taking a particular value (for discrete variables) or falling within a certain range (for continuous variables).

 5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
  - A Cumulative Distribution Function (CDF) gives the probability that a random variable is less than equal to a value, while a Probability Distribution Function (PDF) describes the likelihood of specific values or intervals. CDF is the integral of PDF.

 6. What is a discrete uniform distribution?
  - Discrete Uniform Distribution means having the same probabilty repeated again and again like rolling a dice.

 7. What are the key properties of a Bernoulli distribution?
  - Conditions fro BErnoulli distribution are:
    Number of trial should be finite.
    Each trial should be independent.
    Only two possible outcomes.
 8. What is the binomial distribution, and how is it used in probability?
  - Bernoilli Distributuion is a type of distribution which takes only two values . for example a coin toss.
                    It is used as expeted value in a probabilty distribution.

 9. What is the Poisson distribution and where is it applied?
 - It is a discrete probability distribution that describes the number of events that occur within a fixed interval of time on space given a known average rate of occurance.

 10. What is a continuous uniform distribution?
  - A continuous uniform distribution is a probability distribution where all values within a given range [a,b] are equally likely.

 11. What are the characteristics of a normal distribution?
  - A normal distribution, also known as a Gaussian distribution, has the following key characteristics:
   Bell-Shaped Curve – The distribution is symmetric around the mean.
   Mean = Median = Mode – The central tendency measures are identical.
   Symmetry – The left and right halves of the distribution are mirror images.

 12. What is the standard normal distribution, and why is it important?
  - The standard normal distribution is a normal distribution with mean = 0 and standard deviation = 1. It simplifies probability calculations using z-scores, enables comparisons across datasets, and is essential for hypothesis testing, confidence intervals, and statistical inference.

 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The Central Limit Theorem (CLT) states that, regardless of the original population distribution, the sampling distribution of the sample mean approaches a normal distribution as the sample size increases (n ≥ 30). It's critical because it enables statistical inference, hypothesis testing, and confidence intervals, allowing analysis of real-world data even when the population distribution is unknown.

 14. How does the Central Limit Theorem relate to the normal distribution?
  - The Central Limit Theorem related to normal distributuion with the Standard Normal Distributuion which takes data samples to make large set of data look smaller and simpler, by taking out the assumed mean.

 15. What is the application of Z statistics in hypothesis testing?
  - Z-statistics are used in hypothesis testing to determine if a sample mean significantly differs from a population mean when the population standard deviation is known. It's applied in Z-tests, such as one-sample, two-sample, and proportion tests, for large samples (n ≥ 30) under the normality assumption.

 16. How do you calculate a Z-score, and what does it represent?
  - To calulate Z-score you will have to subtract the assumned mean with the exepcted value('meu') and divide it by variance(sigma).

 17. What are point estimates and interval estimates in statistics?
  - Estimate is a specified observed value used to estimate an unknown population paramater using sample.
                Interval Estimate is a point estimate of a single value that is used to estimate the true value of a opulation parameter.

 18. What is the significance of confidence intervals in statistical analysis?
  - Confidence intervals (CIs) estimate the range where a population parameter (e.g., mean, proportion) likely falls with a given confidence level (e.g., 95%). They provide uncertainty measures in statistical analysis, offering more insight than point estimates. Wider intervals indicate more variability, while narrower intervals suggest higher precision. CIs are crucial in hypothesis testing, decision-making, and predicting population characteristics based on sample data.

 19. What is the relationship between a Z-score and a confidence interval?
  - A Z-score determines the critical value for a confidence interval (CI) in a normal distribution.

 20. How are Z-scores used to compare different distributions?
  - Z-scores standardize data from different distributions by measuring how many standard deviations a value is from the mean. This happens by taking Outliers, Standardization, probabilty calculation, Comparision across Population.

 21. What are the assumptions for applying the Central Limit Theorem?
  - The Central Limit Theorem (CLT) applies under these assumptions:
    Random Sampling – Data must be randomly selected.
    Independence – Observations should be independent of each other.
    Sample Size – A sufficiently large sample (n≥30) is needed for non-normal populations.
    Finite Variance – The population must have a finite standard deviation.
    Underlying Distribution – If the population is normal, CLT holds for any
    n; otherwise, larger samples are required.

 22. What is the concept of expected value in a probability distribution?
  - The expected value (EV) of a probability distribution is the long-term average or mean of a random variable over many trials. It represents the theoretical central value of the distribution.

 23. How does a probability distribution relate to the expected outcome of a random variable?
  - A probability distribution defines the likelihood of different outcomes for a random variable, while the expected outcome (expected value,E(X)) represents the theoretical average of these outcomes over repeated trials.

```python
#2 Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt

low = 1
high = 6
size = 1000

samples = np.random.randint(low, high + 1, size)

values, counts = np.unique(samples, return_counts=True)
pmf = counts / size

plt.stem(values, pmf, basefmt=" ")
plt.title('Discrete Uniform Distribution PMF')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.xticks(values)
plt.show()
```

```python
#3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):

    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        raise ValueError("x must be either 0 or 1 for a Bernoulli distribution.")

p = 0.7
x_values = [0, 1]

for x in x_values:
    print(f"P(X = {x}) = {bernoulli_pdf(x, p)}")
```

```python
#5 Create a Poisson distribution and visualize it using Python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lambda_ = 5
num_samples = 1000

samples = np.random.poisson(lambda_, num_samples)


x_values = np.arange(0, 15)
pmf_values = poisson.pmf(x_values, lambda_)


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(samples, bins=range(0, 16), align='left', rwidth=0.8, color='skyblue', edgecolor='black', density=True)
plt.title(f'Poisson Distribution (λ={lambda_})\nHistogram')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.xticks(range(0, 15))
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.subplot(1, 2, 2)
plt.stem(x_values, pmf_values, basefmt=" ")
plt.title(f'Poisson Distribution (λ={lambda_})\nTheoretical PMF')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.xticks(x_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
#6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt

a = 1
b = 6

x_values = np.arange(a, b + 1)

cdf_values = (x_values - a + 1) / (b - a + 1)

plt.step(x_values, cdf_values, where='post', marker='o', linestyle='-', color='blue')
plt.title(f'CDF of Discrete Uniform Distribution (a={a}, b={b})')
plt.xlabel('x')
plt.ylabel('CDF')
plt.xticks(x_values)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#9 Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):

    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(data, z_scores):

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title('Original Data')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
    plt.title('Z-Scores')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    np.random.seed(42)
    data = np.random.normal(50, 10, 1000)


    z_scores = calculate_z_scores(data)

    plot_z_scores(data, z_scores)
```

```python
#11 Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt

population_mean = 50
population_std_dev = 10
sample_size = 30
num_samples = 10000

sample_means = []
for _ in range(num_samples):
    sample = np.random.normal(population_mean, population_std_dev, sample_size)
    sample_means.append(np.mean(sample))

sample_means_mean = np.mean(sample_means)
sample_means_std_dev = np.std(sample_means)

theoretical_mean = population_mean
theoretical_std_dev = population_std_dev / np.sqrt(sample_size)

plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=50, color='skyblue', edgecolor='black', alpha=0.7, density=True)
plt.title(f'Distribution of Sample Means (CLT)\nSample Size = {sample_size}, Number of Samples = {num_samples}')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)

from scipy.stats import norm
x_values = np.linspace(population_mean - 4 * theoretical_std_dev, population_mean + 4 * theoretical_std_dev, 1000)
pdf_values = norm.pdf(x_values, loc=theoretical_mean, scale=theoretical_std_dev)
plt.plot(x_values, pdf_values, color='red', linewidth=2, label='Theoretical Normal Distribution')

plt.legend()
plt.annotate(f'Sample Means Mean: {sample_means_mean:.2f}\nSample Means Std Dev: {sample_means_std_dev:.2f}',
             xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12, bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
plt.annotate(f'Theoretical Mean: {theoretical_mean:.2f}\nTheoretical Std Dev: {theoretical_std_dev:.2f}',
             xy=(0.05, 0.85), xycoords='axes fraction', fontsize=12, bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.show()
```

```python
#12 Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():

    x_values = np.linspace(-4, 4, 1000)
    pdf_values = norm.pdf(x_values, loc=0, scale=1)

    plt.figure(figsize=(8, 5))
    plt.plot(x_values, pdf_values, color='blue', linewidth=2, label='Standard Normal Distribution')
    plt.title('Standard Normal Distribution (μ=0, σ=1)')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.axvline(x=0, color='red', linestyle='--', label='Mean (μ=0)')
    for std in [-3, -2, -1, 1, 2, 3]:
        plt.axvline(x=std, color='gray', linestyle=':', alpha=0.5, label=f'{std}σ' if std == 1 else None)

    plt.legend()

    plt.show()

plot_standard_normal_distribution()
```

```python
#14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import scipy.stats as stats

def calculate_z_score(data, x):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)

    z_score = (x - mean) / std_dev
    return z_score

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

x = np.random.choice(data)

z = calculate_z_score(data, x)

p_value = stats.norm.cdf(z)

print(f"Dataset Mean: {np.mean(data):.2f}, Standard Deviation: {np.std(data, ddof=1):.2f}")
print(f"Selected Data Point: {x:.2f}")
print(f"Calculated Z-score: {z:.4f}")
print(f"Probability of occurrence (P-value): {p_value:.4f}")

if abs(z) > 2:
    print("The data point is relatively rare (more than 2 standard deviations away from the mean).")
else:
    print("The data point is within a common range (within 2 standard deviations of the mean).")
```

```python
#15 Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
import scipy.stats as stats

np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)

population_mean = 52
population_std_dev = 10

sample_mean = np.mean(sample)
sample_size = len(sample)
sample_std_dev = np.std(sample, ddof=1)

z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

alpha = 0.05

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Population Mean: {population_mean}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis (H0): The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis (H0): No significant difference detected.")

```

```python
#16 Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=30)

sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof=1)
sample_size = len(data)

confidence_level = 0.95
alpha = 1 - confidence_level

z_critical = stats.norm.ppf(1 - alpha/2)

margin_of_error = z_critical * (sample_std_dev / np.sqrt(sample_size))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std_dev:.2f}")
print(f"Confidence Level: {confidence_level * 100}%")
print(f"Z-Critical Value: {z_critical:.4f}")
print(f"Margin of Error: {margin_of_error:.4f}")
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

print("\nInterpretation:")
print(f"We are {confidence_level * 100}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")
```

```python
#19 Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

lambda_value = 5

k_values = np.arange(0, 15)

poisson_cdf = stats.poisson.cdf(k_values, mu=lambda_value)

for k, cdf_value in zip(k_values, poisson_cdf):
    print(f"P(X ≤ {k}) = {cdf_value:.4f}")

plt.figure(figsize=(8, 5))
plt.step(k_values, poisson_cdf, where='mid', label=f'Poisson CDF (λ={lambda_value})', color='blue')

plt.xlabel('Number of occurrences (k)')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) of Poisson Distribution')
plt.legend()
plt.grid()

plt.show()

```

```python
#22 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np
import scipy.stats as stats

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

data_range = np.max(data) - np.min(data)

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
```

```python
#23 Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

mean = np.mean(data)
std_dev = np.std(data, ddof=1)

normalized_data = (data - mean) / std_dev

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color='blue')
plt.axvline(mean, color='red', linestyle='dashed', label=f'Mean={mean:.2f}')
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

plt.subplot(1, 2, 2)
sns.histplot(normalized_data, bins=30, kde=True, color='green')
plt.axvline(0, color='red', linestyle='dashed', label="Mean=0")
plt.title("Z-Score Normalized Data")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()

print(f"Original Mean: {mean:.2f}, Original Std Dev: {std_dev:.2f}")
print(f"Normalized Mean: {np.mean(normalized_data):.2f}, Normalized Std Dev: {np.std(normalized_data, ddof=1):.2f}")
```

```python
#24 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
import scipy.stats as stats

def calculate_skewness_kurtosis(size=1000, mean=50, std_dev=10):

    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std_dev, size=size)


    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data, fisher=True)

    return skewness, kurtosis


skewness, kurtosis = calculate_skewness_kurtosis()


print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurtosis:.4f}")
```
